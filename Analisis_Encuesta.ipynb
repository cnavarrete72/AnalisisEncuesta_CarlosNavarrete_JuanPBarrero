# Ahora, buscaremos si existe algun tipo de agrupamiento en los datos usando k-means

```python
import sklearn
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import average_precision_score
from sklearn.utils.fixes import signature
from sklearn.svm import SVC
from collections import OrderedDict
```

```python
dicc_clases = OrderedDict()
new_data=data.copy()
index=new_data.columns
for i in range(len(new_data.columns)):
        le = LabelEncoder()
        le.fit(new_data[index[i]].astype(str))
        clases=le.classes_
        dicc_clases.update({index[i]:clases})
        new_data[index[i]] = le.fit_transform(new_data[index[i]].astype(str))
```

Buscamos optimizar el numero de ks usando el metodo del codo.

```python
distancias=[]
for i in range(1,10):
    km=KMeans(n_clusters=i)
    km.fit(new_data)
    distancias.append(km.inertia_)

n_cluster=np.linspace(1,len(distancias),len(distancias))
plt.plot(n_cluster,distancias)
plt.scatter(n_cluster,distancias)
plt.xlabel('Num. Clusters')
plt.ylabel('Suma de distancias cuadradas')
```

Nos quedamos con un numero de clusters de 3. Ahora, lo que buscaremos es entender el agrupamiento de estos datos. Para ello, buscaremos las preguntas que generaron los distintos agrupamientos, es decir, aquellas en las que hubo discrepancias entre los encuestados. Esto puede permitir entender las diferentes posturas que existen en la universidad con respecto al plagio y generar politicas con objetivos diferentes para cada poblacion.

```python
kmeans = KMeans(n_clusters=3).fit(new_data)
labels=kmeans.labels_
centroids=kmeans.cluster_centers_
centroids=np.round(centroids)
fil,col=centroids.shape

list_clases=list(dicc_clases.values())
list_nombres=[]
for f in range(fil):
    list_lab=[]
    for c in range(col):
        list_lab.append(list_clases[c][int(centroids[f,c])])
    list_nombres.append(list_lab)
print('Cluster 1:')
print([x for x in list_nombres[0] if x != 'nan'])
print('Cluster 2:')
print([x for x in list_nombres[1] if x != 'nan'])
print('Cluster 3:')
print([x for x in list_nombres[2] if x != 'nan'])
print(dicc_clases.keys())
```

Imprimimos los 3 clusters y los 'ejes', es decir, las preguntas. Miramos por aparte en un excel en donde se muestran las discrepancias (es mas facil visualmente). Y encontramos que:
- En los factores que influyen para que hagan plagio hay diferencias en:
    * La competencia por las notas que se promueve en la universidad
    * Saber que es injusto con mis companeros
- En los factores que influyen en que NO hagan plagio hay diferencias en:
    * Sentir que estoy traicionando la confianza del profesor
- En las actitudes o comportamientos propios de los estudiantes para evitar hacer plagio hay diferencias en:
    * Disminuir el numero de creditos que toman por semestre
    * Entender que cuando hago plagio me engano a mi mismo
- En las actitudes o comportamientos de los profesores para evitar que los estudiantes hagan plagio hay diferencias en:
    * No hacer examenes de memoria o de seleccion multiple
    * Generar panico y sanciones fuertes
- En las actitudes o comportamientos de los companeros para evitar que los estudiantes hagan plagio hay diferencias en
    * Ser menos competitivos
    * Fomentar el companerismo y el estudio en grupo 

Asimismo, parece que hubiera distintas percepciones entre: estudiantes comenzando su carrera, estudiantes mas avanzados y estudiantes a punto de terminar. Consideramos que las diferencias se presentan principalmente entre cada programa, sin embargo, los resultados de este eje en los clusters no son como se esperaban y creemos que para sacar conclusiones sobre la influencia del programa necesitariamos tener mayor cantidad de datos.

#Estudiantes de Fisica
Dentro de las encuestas obtuvimos una alta tasa de respuesta de estudiantes de fisica, esto principalmente porque eran los mas cercanos a nosotros. Vamos a usar esta informacion para entender si existen unos mecanismos especificos que se puedan usar en nuestro departamento para evitar el fraude academico. Para esto, buscaremos predecir la carrera de los estudiantes y veremos la influencia que tiene cada una de las preguntas en determinar si el estudiante es de fisica o no. Fisica es el label 7.

```python
other_data=new_data.copy()
labeled=np.where(other_data["Cual es tu programa principal?"]==7,1,0)
other_data["Cual es tu programa principal?"]=labeled
```

```python
y_train=other_data["Cual es tu programa principal?"]
x_train=other_data.iloc[:,4:] #Nos interesan solo las preguntas.
```

```python
import sklearn
from sklearn.svm import SVC
from sklearn.feature_selection import RFECV

estimator = SVC(kernel="linear")
selector = RFECV(estimator, cv=10,scoring='accuracy')
selector.fit(x_train, y_train)
print('Numero optimo de variables: %d'
 % selector.n_features_)
print(other_data.columns.values[selector.support_])
```

El metodo anterior se denomina 'recursive feature elimination and cross-validated selection' y lo que busca es encontrar las variables que son mas representativas para predecir que un estudiante es de fisica con un Support Vector Classifier. De acuerdo con esto, vemos que son estas 4 preguntas las mas relevantes dentro de los estudiantes de fisica. Creemos que en general los estudiantes consideran que la mejor forma de evitar el plagio es un cuestionamiento moral propio, acompanado de unas sanciones fuertes impuestas por los profesores y la universidad que prevengan estos actos. Cabe destacar tambien que en el departamento parece ser que el fraude es un acto circunstancial y no premeditado, el cual se comete por no preveer las consecuencias en el momento justo de realizarse.

Cabe destacar tambien que al parecer la actitud de los companeros no es del todo relevante para evitar el plagio en Fisica.

```python
#Seleccionamos los mejores predictores usando LinearRegression
from sklearn.linear_model import LinearRegression
estimator2 = LinearRegression()
selector2 = RFECV(estimator2, cv=10)
selector2.fit(x_train, y_train)
print('Numero optimo de variables: %d'
 % selector2.n_features_)
print(other_data.columns.values[selector2.support_])
```

La regresion logistica es consistente con el analisis del fraude como un acto circunstancial.
